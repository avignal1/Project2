```python
# Import statements
import numpy as np
import pandas as pd
from pathlib import Path
%matplotlib inline

#visualizations import statements. Note that not everything was used for the final output
import panel as pn
pn.extension('plotly')
import plotly.express as px
import hvplot.pandas
import matplotlib.pyplot as plt
import holoviews as hv
import os
from dotenv import load_dotenv
```

```python
#Reading csv files for the cleaned data
file_name='data_by_year.csv'
csv_file = Path(f"Data/{file_name}")
data_year_df = pd.read_csv(csv_file)
```

```python
data_year_df.head()
```

```python
year_popularity_df = data_year_df[["year","popularity"]]
```

```python
popularity_scatter = px.scatter(year_popularity_df,
                       x='year',
                       y='popularity',
                       color='popularity',   
                       width=750,
                       height=750,
                       trendline="ols",
                       trendline_color_override="blue",
                       title= "<b>Spotify Popularity Score by Year</b>",
                       hover_name="popularity",
                       color_continuous_scale='dense',
                       hover_data=["popularity"])
popularity_scatter
```

```python
#Reading csv files for the cleaned data
file_name='data_by_genres.csv'
csv_file = Path(f"Data/{file_name}")
data_genres_df = pd.read_csv(csv_file)
```

```python
data_genres_df.head()
```

```python
cleaned_data_genres_df = data_genres_df[["genres","popularity"]]
sorted_genre_popularity_df = cleaned_data_genres_df.sort_values(by=["popularity"],ascending=False)
sorted_genre_popularity_df = sorted_genre_popularity_df.head(100)
```

```python
popularity_genres_bar = px.bar(sorted_genre_popularity_df,
                    x="genres",
                    y="popularity",
                    color="popularity",
                    color_continuous_scale='dense',
                    title= "<b>Popularity by Genres</b>",
                    width=1000,
                     height=600
                    )
                    
popularity_genres_bar
```

```python
#Reading csv files for the cleaned data
file_name='data.csv'
csv_file = Path(f"Data/{file_name}")
data_df = pd.read_csv(csv_file)
```

```python
data_df.head()
```

```python
# Add 3D Plot
import plotly.graph_objects as go

fig_1 = px.scatter_3d(data_genres_df, x='danceability', y='tempo', z='valence',
              color='popularity',hover_name="genres", title="Danceability, Happiness and Tempo on Popularity")

fig_1.show()
```

```python
fig_2 = px.scatter_3d(data_genres_df, x='duration_ms', y='tempo', z='speechiness',
              color='popularity',hover_name="genres", title="Duration, Speechiness and Tempo on Popularity")

fig_2.show()
```

```python
## Insert Dashboard
# Create row
row_1 = pn.Row(popularity_scatter, popularity_genres_bar)
# Create column using Markdown and row object
column = pn.Column(
    '# View of Genre Attributes Impact on Popularity',
    fig_1,
    fig_2)
# Create tabs
tabs = pn.Tabs(
    ("Popularity by Year and Genres", row_1),
    ("Genre Attributes on Popularity", column ))
tabs
```
